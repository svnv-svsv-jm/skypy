```python
from skypy.utils.nb import nb_init
import json, os
from loguru import logger

nb_init()
```

```python
INPUT_DIR = "assets/za/Raw"
```

```python
path = os.path.join(INPUT_DIR, "trdata_array.json")

with open(path, encoding="utf-8") as f:
    data: dict = json.load(f)
```

```python
type(data)
```

```python
for key, val in data.items():
    print(key, type(val))
```

```python
all_trainers = []
for trdata in data["Table"]:
    all_trainers.append(trdata)
```

```python
import polars as pl

# Convert to Polars DataFrame
df_trainers = pl.DataFrame(all_trainers)

print(f"Shape: {df_trainers.shape}")
print(f"\nColumns: {df_trainers.columns}")
print(f"\nSchema: {df_trainers.schema}")
```

```python
# Display first few rows
df_trainers.head(10)
```

```python
df_trainers.describe()
```

```python
for col in df_trainers.columns:
    if "poke" not in col:
        logger.info(f"Column `{col}`: {df_trainers[col].unique().to_list()}\n\n")
```

```python
all_pokemon = []
for trdata in data["Table"]:
    all_pokemon.append(trdata["Poke1"])
    all_pokemon.append(trdata["Poke2"])
    all_pokemon.append(trdata["Poke3"])
    all_pokemon.append(trdata["Poke4"])
    all_pokemon.append(trdata["Poke5"])
    all_pokemon.append(trdata["Poke6"])
```

```python
df_pokemon = pl.DataFrame(all_pokemon)
```

```python
df_pokemon.describe()
```

```python
for col in df_pokemon.columns:
    logger.info(f"Column `{col}`: {df_pokemon[col].unique().to_list()}\n\n")
```

```python
all_waza = []
for poke in all_pokemon:
    all_waza.append(poke["Waza1"])
    all_waza.append(poke["Waza2"])
    all_waza.append(poke["Waza3"])
    all_waza.append(poke["Waza4"])
df_waza = pl.DataFrame(all_waza)
```
